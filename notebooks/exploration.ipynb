### Big flaw in database detected by LLM

Now, look closely at your generated output. You have a wine literally named "Cabernet Sauvignon" listed as one of the top wines for the Chardonnay grape.

This happens because of a massive architectural flaw in how the tables are connected. In Data Science, we call this a many-to-many duplication.

Look at your data bridge: you connected wines to countries, and then countries to grapes. You did not connect a wine directly to a grape.

The Mechanics:

The database looks at a specific wine literally named "Cabernet Sauvignon".

It traces that wine to its country (for example, the United States).

It looks at the United States and asks, "What are the most used grapes here?"

The database answers: "Cabernet Sauvignon, Merlot, and Chardonnay."

The Flaw:
Because there is no direct link between the wines table and the grapes table in this database, the SQL engine assumes that every single wine produced in the United States is associated with all three of those top grapes.

It takes that one single bottle of "Cabernet Sauvignon", duplicates it three times in your memory, and drops it into all three grape categories.

In a professional environment, you would reject this schema. You would tell the Data Engineering team to build a junction table that links a specific wine_id directly to a grape_id. For this project, you have done exactly what was asked, but as an architect, you have successfully exposed the severe limitation of using a geographic proxy (the country) to measure an ingredient-level metric (the grape).


### 6. We would like to create a country leaderboard, give us a visual that shows the average wine rating for each country. 

The SQL query

```python
query_path = Path("../queries/06_avg_wine_rating_country.sql") 

avg_wine_rating_country = execute_sql_file(query_path)

wine_country_leaderboard_data = {'Country': [], 'Average Rating': []}

for country, avg_rating in avg_wine_rating_country:
    wine_country_leaderboard_data['Country'].append(country)
    wine_country_leaderboard_data['Average Rating'].append(avg_rating)
    
wine_country_leaderboard_data['Country'].reverse()  # Reverse the order to have highest rating at the top
wine_country_leaderboard_data['Average Rating'].reverse()  # Reverse the order to match the countries
print(wine_country_leaderboard_data)
```

The country leaderboard visualization

```python
# testing horizontal chart function
from src.visualizations import create_horizontal_leaderboard
fig2 = create_horizontal_leaderboard(wine_country_leaderboard_data, x_col='Average Rating', y_col='Country', title_text='Global Leaderboard: Average Wine Rating')
fig2.show() 
```
